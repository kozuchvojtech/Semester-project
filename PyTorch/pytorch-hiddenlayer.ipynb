```python
# https://pypi.org/project/hiddenlayer/
# https://github.com/waleedka/hiddenlayer

import hiddenlayer as hl
import numpy as np
import torch
```

```python
x_train = np.array([[1.7], [2.5], [5.5], [7.9], [8.8], [2.4], [2.4], [8.89], [5], [4.4]], dtype=np.float32)
y_train = np.array([[1.9], [2.68], [4.22], [8.19], [9.69], [3.4], [2.6], [8.8], [5.6], [4.7]], dtype=np.float32)
```

```python
X_train = torch.tensor(x_train)
Y_train = torch.tensor(y_train)

# input feature
inp = 1
# output value
out = 1

# hidden layer contains 100 neurons
hid = 100
```

```python
# sequential network with two linear layers
model1 = torch.nn.Sequential(torch.nn.Linear(inp, hid), 
                             torch.nn.Linear(hid, out))
```

```python
# torch.zeros -> test data of the expected dimensions
hl.build_graph(model1, torch.zeros([10,1]))
```

```python
# previous example 10 inputs; now only one => detailed visualization -> we can see all of the individual operations performed in the linear layers
# each of the layers: transpose of the matrix, its multiplication and adding
hl.build_graph(model1, torch.zeros([1]))
```

```python
model2 = torch.nn.Sequential(torch.nn.Linear(inp, hid),
                             torch.nn.Linear(hid, hid),
                             torch.nn.Sigmoid(),
                             torch.nn.Linear(hid, out))
```

```python
hl.build_graph(model2, torch.zeros([10,1]))
```

```python
hl.build_graph(model2, torch.zeros([1]))
```
